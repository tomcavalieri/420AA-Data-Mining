```python
df = pd.read_csv("/kaggle/input/cleaned-credit/cleaned_credit.csv")
df
```

```python
#since:
#1.)Number of credit problems
#2.)Term
#3.)Purpose
#4.)Loan Status
#are unbalanced I will delete them
#and do classification without them (For 'Loan_Status I will do sampling methods')

continuous = ['Current_Loan_Amount', 'Annual_Income', 'Monthly_Debt', 'Current_Credit_Balance', 'Maximum_Open_Credit', 'Years_of_Credit_History']

discrete= ['Number_of_Credit_Problems','Number_of_Open_Accounts','Credit_Score']

categorical= ['Term','Home_Ownership','Years_in_current_job','Purpose']

target=['Loan_Status']
```

df=df.drop(['Number_of_Credit_Problems'],axis=1)
df=df.drop(['Term'],axis=1)
df=df.drop(['Purpose'],axis=1)

```python
y=pd.DataFrame(df['Loan_Status'])
y
```

```python
x=df.drop(['Loan_Status'],axis=1)
x
```

```python
from collections import Counter

counter=Counter(y['Loan_Status'])
counter
```

```python
import matplotlib.pyplot as plt

plt.figure()
y['Loan_Status'].value_counts().plot(kind='bar', title='Loan_Status')
```

```python
from sklearn.preprocessing import LabelEncoder 

label_encoder = LabelEncoder()

y = pd.DataFrame(label_encoder.fit_transform(y))
y.columns = ['Loan_Status']
y
```

```python
from imblearn.over_sampling import RandomOverSampler

oversample = RandomOverSampler(sampling_strategy='minority',random_state=0)
x_over, y_over = oversample.fit_resample(x, y)
y_over
```

```python
from imblearn.under_sampling import RandomUnderSampler

undersample = RandomUnderSampler(sampling_strategy='majority',random_state=0)
x_under, y_under = undersample.fit_resample(x, y)
y_under
```

```python
dummies=pd.get_dummies(x,columns=categorical)
dummies
```

```python
from imblearn.over_sampling import SMOTE

smote=SMOTE(random_state=0)

x_smote,y_smote=smote.fit_resample(dummies,y)
x_smote
```

```python
from imblearn.over_sampling import ADASYN 

adasyn=ADASYN(random_state=0)

x_adasyn,y_adasyn=adasyn.fit_resample(dummies,y)
x_adasyn
```

```python
print("Imbalanced data:",np.shape(x),"target:",Counter(y['Loan_Status']))
print("Oversampled data:",np.shape(x_over),"target:",Counter(y_over['Loan_Status']))
print("Undersampled data:",np.shape(x_under),"target:",Counter(y_under['Loan_Status']))
print("SMOTED data:",np.shape(x_smote),"target:",Counter(y_smote['Loan_Status']))
print("ADASYNED data:",np.shape(x_adasyn),"target:",Counter(y_adasyn['Loan_Status']))
```

```python
dummies=pd.get_dummies(df,columns=categorical)
dummies
```

```python
dummies_over=pd.get_dummies(x_over,columns=categorical)
dummies_over
```

```python
dummies_under=pd.get_dummies(x_under,columns=categorical)
dummies_under
```

```python
df=dummies.drop(['Loan_Status'],axis=1)
df
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df, y, test_size=0.2,shuffle=True,random_state=0)
x_train_over, x_test_over, y_train_over, y_test_over = train_test_split(dummies_over, y_over, test_size=0.2,shuffle=True,random_state=0)
x_train_under, x_test_under, y_train_under, y_test_under = train_test_split(dummies_under, y_under, test_size=0.2,shuffle=True,random_state=0)
x_train_smote, x_test_smote, y_train_smote, y_test_smote = train_test_split(x_smote, y_smote, test_size=0.2,shuffle=True,random_state=0)
x_train_adasyn, x_test_adasyn, y_train_adasyn, y_test_adasyn = train_test_split(x_adasyn, y_adasyn, test_size=0.2,shuffle=True,random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x_train_norm=scaler.fit_transform(x_train)
x_test_norm=scaler.fit_transform(x_test)

x_train_norm_over=scaler.fit_transform(x_train_over)
x_test_norm_over=scaler.fit_transform(x_test_over)

x_train_norm_under=scaler.fit_transform(x_train_under)
x_test_norm_under=scaler.fit_transform(x_test_under)

x_train_norm_smote=scaler.fit_transform(x_train_smote)
x_test_norm_smote=scaler.fit_transform(x_test_smote)

x_train_norm_adasyn=scaler.fit_transform(x_train_adasyn)
x_test_norm_adasyn=scaler.fit_transform(x_test_adasyn)
```

```python
#x_train = pd.DataFrame(x_train, index=df.index, columns=df.columns)
x_train_norm=pd.DataFrame(x_train, index=x_train.index, columns=x_train.columns)
x_test_norm=pd.DataFrame(x_test, index=x_test.index, columns=x_test.columns)

x_train_norm_over=pd.DataFrame(x_train_over, index=x_train_over.index, columns=x_train_over.columns)
x_test_norm_over=pd.DataFrame(x_test_over, index=x_test_over.index, columns=x_test_over.columns)

x_train_norm_under=pd.DataFrame(x_train_under, index=x_train_under.index, columns=x_train_under.columns)
x_test_norm_under=pd.DataFrame(x_test_under, index=x_test_under.index, columns=x_test_under.columns)

x_train_norm_smote=pd.DataFrame(x_train_smote, index=x_train_smote.index, columns=x_train_smote.columns)
x_test_norm_smote=pd.DataFrame(x_test_smote, index=x_test_smote.index, columns=x_test_smote.columns)

x_train_norm_adasyn=pd.DataFrame(x_train_adasyn, index=x_train_adasyn.index, columns=x_train_adasyn.columns)
x_test_norm_adasyn=pd.DataFrame(x_test_adasyn, index=x_test_adasyn.index, columns=x_test_adasyn.columns)
```

```python
#NORMAL DATASET


from sklearn import tree
from sklearn import metrics
from sklearn.model_selection import GridSearchCV

decisionTree = tree.DecisionTreeClassifier(random_state=0)

#grid search
criterion = ['gini','entropy']
max_depth = np.arange(1,10)
splitter=['best','random']
min_impurity_decrease=[1e-16,1e-14]
max_features=[None,'auto','sqrt']

param_grid = dict(criterion=criterion, max_depth=max_depth,splitter=splitter,min_impurity_decrease=min_impurity_decrease,
                 max_features=max_features)
grid = GridSearchCV(estimator=decisionTree, param_grid=param_grid, n_jobs=-1, cv=3,return_train_score=True)

grid_result = grid.fit(x_train_norm, y_train)

print("Best result is: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```
